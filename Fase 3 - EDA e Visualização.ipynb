# Fase 3 - EDA e Visualização

```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
% matplotlib inline
```

```python
entrada = pd.read_csv('saidaPreProcessamento.csv',header = 0, sep=',',encoding='utf-8',index_col=0)

entrada['Nome'] = entrada.Nome.astype('category')
entrada['Nota'] = entrada.iloc[:,1:5].mean(axis=1)
entrada.sort_values(by='Nota',inplace=True,ascending=False)
#entrada.dtypes
#avaliacoes = entrada.iloc[:,0:5].groupby('Nome')

avaliacoes = entrada.iloc[:,0:5]
avaliacoes['Nota'] = entrada.Nota
#avaliacoes.head()

avaliacoes.set_index('Nome',inplace=True)
primeiros10nomes = entrada.iloc[:10,:].Nome.values
avaliacoes.loc[primeiros10nomes,['Nota']].plot(kind='bar',figsize=(20,8))
avaliacoes.loc[primeiros10nomes,['localizacao','quartos','facilidades']].plot(kind='bar',figsize=(20,8))
```
